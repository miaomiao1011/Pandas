# Add and Drop Columns

```python
df['Total']=df['Physical_effects']+df['Multimedia']+df['ET_contact']+df['Abduction']
df.head()
```

```python
df.head()
```

```python
df['Total2']=df.iloc[:,3:7].sum(axis=1)
df.head()
```

```python
df['Total3']=df.loc[:,"Physical_effects":"Abduction"].sum(axis=1)
df.head()
```

```python
df=df.drop(columns=['Total2','Total3'])
df.head()
```

# Rearrage Columns

```python
cols=list(df.columns.values)
cols
```

```python
df=df[cols[0:3]+[cols[-1]]+cols[3:7]]
df.head()
```
